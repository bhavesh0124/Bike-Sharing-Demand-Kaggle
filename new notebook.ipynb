```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df_train=pd.read_csv("train.csv")
df_test=pd.read_csv("test.csv")
```

```python
df_train.head(2)
```

```python
df_train.hist(bins=50,figsize=(20,15))
```

```python
datetime = df_train['datetime']
train_hour = []
for i in datetime:
    data = i.split(' ')
    time = data[1]
    new_data = time.split(':')
    train_hour.append(new_data[0])
df_train['hour'] = train_hour
```

```python
plt.plot(df_train['hour'], df_train['count'])
plt.xlabel('hour')
plt.ylabel('count')
```

```python
import datetime
train_weekday = []
test_weekday = []

for i in df_train['datetime']:
    train_weekday.append(datetime.datetime.strptime(i, "%Y-%m-%d %H:%M:%S").strftime("%A"))
    
for i in df_test['datetime']:
    test_weekday.append(datetime.datetime.strptime(i, "%Y-%m-%d %H:%M:%S").strftime("%A"))

df_train['weekday'] = train_weekday
df_test['weekday'] = test_weekday
```
